```python
def freq_lettres(texte):
  d = {}
  texte = texte.lower()
  for i in texte:
    if i != " ":
      d[i] = d.get(i,0) + 1
  return d
```

```python
t = freq_lettres("Bonjour bonjour")
assert t["o"] == 4 and t["b"] == 2 and " " not in t
```

```python
def freq_paires(texte):
  d={}
  texte = texte.lower()
  for a,b in zip(texte, texte[1:]):
    m = (a,b)
    d[m] = d.get(m,0) + 1
  return d
```

```python
p = freq_paires("abab")
assert p[("a","b")] == 2 and p[("b","a")] == 1
```

```python
p = freq_paires("abab")
assert p[("a","b")] == 2 and p[("b","a")] == 1
```

```python
def top_n(d,n):
  L = sorted(d.items(), key  = lambda valeur: valeur[1], reverse = True)
  M = [clef for clef,val in L[:n]]
  return (M)


```

```python
assert top_n({"a":3,"b":9,"c":5}, 2) == ["b","c"]
```

```python
def normalise(L):
  total  =sum(L)
  M = [i/total for i in L]
  return M
```

```python
n = normalise([2, 2, 4])
assert abs(sum(n) - 1) < 1e-9 and abs(n[2] - 0.5) < 1e-9
```

```python
def moyenne(L):
  return sum(L)/len(L)
```

```python
def variance(L):
  m = moyenne(L)
  prem = [(i-m)**2 for i in L]
  return moyenne(prem)
```

```python
assert moyenne([1,2,3,4]) == 2.5 and abs(variance([1,2,3,4]) - 1.25) < 1e-9
```

```python
def mots_frequents(texte,n):
  texte = texte.lower()
  mots = texte.split()
  d = {}
  for mot in mots:
    d[mot] = d.get(mot,0) +1
  L = sorted(d.items(), key = lambda x: x[1], reverse = True)
  M = [clef for clef,val in L[:n]]
  return M
```

```python
assert mots_frequents("le chat le chien le chat", 1) == ["le"]
```

```python
def matmul(A,B):
  m = len(A[0])
  n = len(A[1])
  p = len(B[1])
  C = [[0 for _ in range(p)] for _ in range(m)]
  for i in range(m):
    for j in range(p):
      for k in range(n):
        C[i][j] += A[i][k]*B[k][j]
  return C
```

```python
C = matmul([[1,2],[3,4]], [[5,6],[7,8]])
assert C == [[19,22],[43,50]]
```

```python
import time
```

```python
def chrono(f,*args):
  debut = time.perf_counter()
  resultat = f(*args)
  fin = time.perf_counter()
  duree = fin-debut
  return resultat, duree
```

```python
def carre(x):
  return(x**2)
```

```python
print(chrono(carre,5))
```
